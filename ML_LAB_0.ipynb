```python
import numpy as np
import pandas as pd
import seaborn as sns
import re
sns.set()
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-02-11/hotels.csv")
```

```python
print(df.head())
```

```python
df = df.reset_index().rename(columns={'index':'id'})
print(df)
```

```python
def check_values(df):
  data = []
  for column in df.columns:
     data.append([ column,df[column].dtype,df[column].isna().sum(),round(100*(df[column].isna().sum()/len(df)),2),df[column].nunique() ])
  return pd.DataFrame(columns=['Data_Features','Data_Type','Null','Null_Percentage','Unique_Value'], data=data)

```

```python
check_values(df)
```
